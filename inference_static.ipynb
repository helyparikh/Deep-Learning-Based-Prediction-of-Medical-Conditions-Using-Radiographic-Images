```python
import os
import random
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import numpy as np
import imutils
import warnings
warnings.filterwarnings('ignore')
```

```python
from keras.models import load_model
 
# load model
model = load_model('resnet_backup.h5')
```

```python
columns = ['Atelectasis', 'Cardiomegaly', 'Consolidation', 'Edema', 'Enlarged', 'Fracture', 'Lung Lesion',
           'Lung Opacity', 'No Finding', 'Pleural Effusion', 'Pleural Other', 'Pneumonia', 'Pneumothorax',
          'Support Devices']
```

## Please provide image path here to get prediction result

```python
imagepath= 'output_5_0.png'
```

```python
image = cv2.imread(imagepath)
plt.imshow(image)
image = cv2.resize(image, (224,224))
image = image / 255
prediction = model.predict(np.asarray([image]))
```

```python
for i in range(len(prediction[0])):
    print(columns[i] +": "+str(prediction[0][i]))
```
